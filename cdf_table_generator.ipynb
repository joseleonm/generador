```
### TABLE PARAMS ###
verticalMin = -1.8
horizontalMin = 0.03
numRows = 3
numCols = 3

### CREATE DATA FOR TABLE ###
verticalMin = float(verticalMin)
horizontalMin = float(horizontalMin)

vertical_floats = [verticalMin + n*0.1 for n in range(numRows)]
horizontal_floats = [horizontalMin + n*0.01 for n in range(numCols)]

def includeDecimalPlaces(number, numDecimalPlaces):
    result = str(round(number, numDecimalPlaces)) + '0'*numDecimalPlaces
    try:
        idx = result.index('.')
        return result[:idx + numDecimalPlaces + 1]
    except:
        raise Exception('{} is not a valid input'.format(z))

vertical = [includeDecimalPlaces(n, 1) for n in vertical_floats]
horizontal = [includeDecimalPlaces(n, 2) for n in horizontal_floats]

import scipy.stats as st
arr = [[0 for _ in horizontal] for _ in vertical]
for i, tenths in enumerate(vertical_floats):
    for j, hundredths in enumerate(horizontal_floats):        
        if tenths >= 0:
            z = tenths + hundredths
        else:
            z = tenths - hundredths

        phi = st.norm.cdf(z)
        arr[i][j] = includeDecimalPlaces(phi, 4)

### CREATE TABLE ###

tableTagBegin = '<table rules=all cellpadding=5 align="center" style="border:1px solid black; margin: auto;">' + '\n'
tableTagEnd = '</table>'
tableHeader = '<tr> <th bgcolor="#D6EAF8" style="text-align:center" width=25%>$z$</th>'\
              + ''.join(['<th bgcolor="#D6EAF8" style="text-align:center">$\, {} \,$</th>'.format(z) for z in horizontal]) \
              + '\n'
tableRows = ''
for i, row in enumerate(arr):
    rowString = '<tr> <th bgcolor="#D6EAF8" style="text-align:center">$\, {} \,$</th>'.format(vertical[i])
    for phi in row:
        rowString += '<td style="text-align:center" > $\, {} \,$ </td>'.format(phi)
    rowString += '</tr>\n'
    tableRows += rowString

html = tableTagBegin + tableHeader + tableRows + tableTagEnd
print(html)
```
